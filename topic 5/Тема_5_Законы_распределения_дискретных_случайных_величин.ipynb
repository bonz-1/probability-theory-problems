## **Биномиальный закон распределения**

### **Задача**

IT-компания проводит автоматическое обновление мобильного приложения. При каждом обновлении вероятность того, что оно установится без ошибок, равна 0.85. За ночь система выполняет обновление на 8 случайно выбранных устройствах. Составить закон распределения количества устройств, на которых обновление прошло успешно. Найти математическое ожидание, дисперсию, среднее квадратичное отклонение и моду, а также построить многоугольник распределения.

```python
from scipy.stats import binom
```

```python
n, p = 8, 0.85
X = np.arange(n + 1)

binomial_dist = binom(n, p)
probabilities = binomial_dist.pmf(X)

df = pd.DataFrame(
    data=probabilities.reshape(1, -1),
    columns=X,
    index=['p']
)

df
```

```python
mode = np.array([X[np.argmax(probabilities)]])

print(f'Математическое ожидание = {binomial_dist.mean():.4f}')
print(f'Дисперсия = {binomial_dist.var():.4f}')
print(f'Среднее квадратичное отклонение = {binomial_dist.std():.4f}')
print(f'Мода = {mode}')
```

```python
plt.figure(figsize=(8, 3))
plt.plot(X, probabilities, linewidth=2, marker='s', label='Многоугольник распределения')
plt.title(f'Биномиальное распределение n = {n}, p = {p}')
plt.xlabel('m — количество успешных обновлений')
plt.ylabel('P(m)')
plt.xticks(X)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

## **Распределение Пуассона**

### **Задача**

В дата-центре работают 20 000 серверов. Каждый сервер за час с вероятностью 0.0003 может выйти из строя из-за мелкой аппаратной неисправности. Требуется составить закон распределения количества вышедших из строя серверов за один час. Найти математическое ожидание, дисперсию, среднее квадратичное отклонение и моду, а также построить многоугольник распределения.

```python
from scipy.stats import poisson
```

```python
lam = 6
X = np.arange(16)

poisson_dist = poisson(lam)
probabilities = poisson_dist.pmf(X)

df = pd.DataFrame(
    data=probabilities.reshape(1, -1),
    columns=X,
    index=['p']
)

df
```

```python
mode = np.array([X[np.argmax(probabilities)]])

print(f'Математическое ожидание = {poisson_dist.mean():.4f}')
print(f'Дисперсия = {poisson_dist.var():.4f}')
print(f'Среднее квадратичное отклонение = {poisson_dist.std():.4f}')
print(f'Мода = {mode}')
```

```python
plt.figure(figsize=(8, 3))
plt.plot(X, probabilities, linewidth=2, marker='s', label='Многоугольник распределения')
plt.title(f'Распределение Пуассона λ = {lam}')
plt.xlabel('x — количество отказавших серверов')
plt.ylabel('P(x)')
plt.xticks(X)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

## **Геометрическое распределение**

### **Задача**

В компании по разработке дронов тестируют новую модель. Каждый запуск дрона с вероятностью 0.1 заканчивается аварийным сбоем. Составить закон распределения числа запусков до сбоя и отправки на диагностику. Найти математическое ожидание, дисперсию, среднее квадратичное отклонение и моду, а также построить многоугольник распределения.

```python
from scipy.stats import geom
```

```python
p = 0.1
X = np.arange(1, 21)

geom_dist = geom(p)
probabilities = geom_dist.pmf(X)

df = pd.DataFrame(
    data=probabilities.reshape(1, -1),
    columns=X,
    index=['p']
)

df
```

```python
mode = np.array([1])

print(f'Математическое ожидание = {geom_dist.mean():.4f}')
print(f'Дисперсия = {geom_dist.var():.4f}')
print(f'Среднее квадратичное отклонение = {geom_dist.std():.4f}')
print(f'Мода = {mode}')
```

```python
plt.figure(figsize=(8, 3))
plt.plot(X, probabilities, linewidth=2, marker='s', label='Многоугольник распределения')
plt.title(f'Геометрическое распределение p = {p}')
plt.xlabel('x — номер запуска с первым сбоем')
plt.ylabel('P(x)')
plt.xticks(X)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

## **Гипергеометрическое распределение**

### **Задача**

Компания по тестированию электроники получила партию из 50 смартфонов, из которых 10 имеют дефектные аккумуляторы. Инженер случайным образом выбирает 8 смартфонов для проверки. Требуется составить закон распределения количества дефектных смартфонов среди выбранных. Найти математическое ожидание, дисперсию, среднее квадратичное отклонение и моду, а также построить многоугольник распределения.

```python
from scipy.stats import hypergeom
```

```python
M = 50
K = 10
n = 8
X = np.arange(0, n+1)

hg_dist = hypergeom(M, K, n)
probabilities = hg_dist.pmf(X)

df = pd.DataFrame(
    data=probabilities.reshape(1, -1),
    columns=X,
    index=['p']
)

df
```

```python
mode = X[np.argmax(probabilities)]

print(f'Математическое ожидание = {hg_dist.mean():.4f}')
print(f'Дисперсия = {hg_dist.var():.4f}')
print(f'Среднее квадратичное отклонение = {hg_dist.std():.4f}')
print(f'Мода = {mode}')
```

```python
plt.figure(figsize=(8, 3))
plt.plot(X, probabilities, linewidth=2, marker='s', label='Многоугольник распределения')
plt.title(f'Гипергеометрическое распределение M={M}, K={K}, n={n}')
plt.xlabel('x — количество дефектных смартфонов')
plt.ylabel('P(x)')
plt.xticks(X)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
